## Description:
This housing dataset aims to predict land prices according to user preference. The datasets consists of several variables which includes POSTED_BY, UNDER_CONSTRUCTION, RERA, BHK_NO., BHK_OR_RK, SQUARE_FT, READY_TO_MOVE, RESALE, ADDRESS, LONGITUDE, LATITUDE, TARGET(PRICE_IN_LACS).

Dataset URL: https://www.kaggle.com/datasets/anmolkumar/house-price-prediction-challenge/data?select=train.csv

### Name of the columns with their meanings

1) Posted_By --> Who posted the Deal<br> 
2) Under_Construction --> Is the Apartment build Or Under Construction <br> 
3) Rera --> Real Estate Regulatory Authority<br> 
4) Bhk_No. --> Bedroom Hall Kitchen_Number<br> 
5) Bhk_or_Rk --> Bedroom Hall Kitchen or Room Kitchen<br> 
6) Square_Ft --> It's defined as a square with sides that are 1 foot long<br>
7) Ready_To_Move --> Is it available for occupancy or not<br> 
8) Resale --> It refers to selling a previously owned property<br> 
9) Address --> its is the specific location identifier of a property<br> 
10) Longitude --> It is a geographic coordinate that specifies the east-west position of a point<br>
11) Latitude --> It is a geographic coordinate that specifies the north-south position of a point<br>
12) Target(Price_in_lacs) --> desired or expected sale/purchase value of the property expressed in lakhs<br>

By calculating correlation coefficients, such as Pearson's correlation coefficient for continuous variables or rank-based correlations for non-linear or ordinal data, data scientists can identify patterns and dependencies in the data. ## 7) Apply Feature Scalling

Robust Scaling is similar to standard scaling, but it uses the median and interquartile range (IQR) instead of the mean and standard deviation. It can help to:<br>1) 
Reduce the effect of outlier<br>
2) 
Improve the model's ability to capture relationships between features

```python
# Apply Standard Scaler
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
X_scaled = scaler.fit_transform(X)
```

## 8) Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=75)
```


X is your feature data (e.g., a Pandas DataFrame or a NumPy array) <br>
y is your target variable (e.g., a Pandas Series or a NumPy array) <br>
test_size=0.2 means that 20% of the data will be used for testing, and the remaining 70% will be used for training <br>
random_state=75 ensures that the split is reproducible (i.e., you'll get the same split every time you run the code) <br>
#### The train_test_split function returns four arrays:
X_train: the training feature data  <br>
X_test: the testing feature data <br>
y_train: the training target variable <br>
y_test: the testing target variable <br>
You can then use the training data to train your machine learning model, and the testing data to evaluate its performance.
